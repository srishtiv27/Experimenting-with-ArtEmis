```python
WIKI_IMAGE_ART_DIRECTORY = '/content/drive/MyDrive/CV/ArtEmis/images/wikiart'

import random
from PIL import Image
import matplotlib.pyplot as plt

def show_random_image_and_caption():
  random_index = random.randint(0, len(artemis_unique))
  random_observation = artemis_unique.iloc[random_index]
  image_path = WIKI_IMAGE_ART_DIRECTORY + '/' + random_observation['art_style'] + '/' + random_observation['painting'] + '.jpg'
  image = Image.open(image_path)
  plt.imshow(image)
  for i in range(len(artemis_preprocessed)):
    if artemis_preprocessed.iloc[i]['painting'] == random_observation['painting']:
      print('Emotion: ', artemis_preprocessed.iloc[i]['emotion'])
      print('Caption: ', artemis_preprocessed.iloc[i]['utterance'])
      print()
```

```python
show_random_image_and_caption()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 10))

def show_percentage(percentage):
  return str(round(percentage, 2)) + "%"

colors = sns.color_palette('pastel')[0:9]

# Drawing a pie chart, and depicting the number of males and females in it
artemis_preprocessed.groupby('emotion').size().plot(kind='pie', textprops={'fontsize': 15}, autopct=show_percentage, colors=colors, ylabel = '')


plt.title('Distribution of paintings by dominant emotion(Percentage)', fontsize= 20)

plt.show()
```

```python
plt.figure(figsize=(15, 10))
artemis_preprocessed.groupby('emotion').size().plot(kind='bar', color= colors)
plt.ylabel('Count')
plt.title('Distribution of paintings by dominant emotion(Count)')
plt.show()
```

```python
# Function to group all the utterances corresponding to a particular emotion.
def group_by_emotion(emotion):
    utterances_by_emotion = artemis_preprocessed.loc[artemis_preprocessed['emotion'] == emotion]
    return ' '.join(utterance for utterance in utterances_by_emotion['utterance'])
```

```python
from wordcloud import WordCloud, STOPWORDS

# Function to make a wordcloud given utterance corresponding to an emotion.
def make_wordcloud(emotion_utterances):
    return WordCloud(
                stopwords = set(STOPWORDS),
                background_color = 'white').generate(emotion_utterances)
```

```python
# Drawing wordcloud for awe emotion.
plt.figure(figsize=(12,24))
wordcloud_awe = make_wordcloud(group_by_emotion('awe'))
plt.imshow(wordcloud_awe)
plt.show()
```
